```python
import pandas as pd
import requests
import json
import datetime
import time
```

```python
def scrape_api(url):
    # Send an HTTP GET request to the API endpoint
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON data from the response
        data = response.json()

        # Print a limited portion of the JSON response for inspection
#         print(json.dumps(data['response']['docs'][:5], indent=4))  # Prints the first 5 documents
    else:
        print(f"Request failed with status code: {response.status_code}")
        
    return data  
```

```python
def df_creator(apilist):
    headlines=[]
    date1=[]
    time1=[]
    # Store data into the fields 
    for item in apilist['response']['docs']:
        headlines.append(item['abstract'])
    #     date_and_time=item['pub_date']
        date1.append(item['pub_date'][:10])
        time1.append(item['pub_date'][10:])
    #     ratingScore.append(item['leftView']['rateScore'])
    #     quantity.append(item['rightView']['text'].split(' Sold')[0]
#     print(headlines)
#     print(date)
#     print(time)
    #Combine all the columns together
    output=pd.DataFrame({'Date':date1,'Time':time1,'Headlines':headlines})
    output
    return output
```

```python
def df_concat(datfra1,datfra2):
    
    final_data = pd.concat([datfra1, datfra2], ignore_index=True, sort=False)
    return final_data
```

```python
# Example usage
def required_time(From_year, To_year, from_month=1, to_month=12):
    year=From_year
    month=from_month
    news_data=pd.DataFrame()
    # print(dt['response']['docs'])

    for i in range(From_year,To_year+1):
        month=from_month
        for j in range(from_month-1,to_month):
        
            print(year,month)
            api_url = 'https://api.nytimes.com/svc/archive/v1/'+str(year)+'/'+str(month)+'.json?api-key=1InslIO2svxp9asyHPStTM1dAWKdolNl'  # Replace with the actual API endpoint URL
            dt=scrape_api(api_url)
            
            df1=df_creator(dt)
            
            # Use pandas.concat() method to ignore_index 
            news_data = df_concat(news_data,df1)
            # print(month)
            month+=1
            time.sleep(15)
            
        year+=1


    return news_data
```
